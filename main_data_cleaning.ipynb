# Now we make the final merge


# This is the combination of 


* Committee Master
    * gives you all the names of the committee
    * gives you CONNECTED_ORG_NM, which is very important and tells you the organizations connected to the commitee
* Candidate Master
    * Links the candidate ID to the candidate name and election year and state and city

* contributions_from_committees_to_cand
    * this tells you the individual transactions between committees and candidates
